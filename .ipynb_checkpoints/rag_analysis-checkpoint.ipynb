# Docshunter RAG Benchmarks Analysis

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from typing import Dict, List
import json
```

```python
sns.set_theme()
sns.set_palette("husl")
```

```python
with open('final_results_bge-large-en-v1.5.json', 'r') as f:
    json_data = json.load(f)

results_data = {
    "test_parameters": {
        "chunk_sizes": [500, 800, 1000, 1500, 2000],
        "strategies": ["regex", "recursive_text_splitter"],
        "context_retrieval": [True, False]
    },
    "all_results": json_data
}
```

# Metrics Distribution

```python
plt.figure(figsize=(12, 6))
metrics = ['faithfulness', 'answer_relevancy', 'answer_similarity', 'answer_correctness']
sns.boxplot(data=df[metrics])
plt.title('Distribution of all Metrics using all configurations')
plt.xticks(rotation=45)
plt.ylabel('Score')
plt.show()
```

# Impacto of Chunk Size on Each Metric

```python
plt.figure(figsize=(15, 10))
for i, metric in enumerate(metrics, 1):
    plt.subplot(2, 2, i)
    sns.lineplot(
        data=df,
        x='chunk_size',
        y=metric,
        hue='strategy',
        style='context_retrieval',
        markers=True,
        marker='o'
    )
    plt.title(f'Impact of Chunk Size on {metric.title()}')
plt.tight_layout()
plt.show()
```

# Strategy Comparison


```python
plt.figure(figsize=(12, 6))
strategy_comparison = df.groupby('strategy')[metrics].mean()
strategy_comparison.plot(kind='bar')
plt.title('Average Metrics usingText Splitting Strategy')
plt.xticks(rotation=45)
plt.legend(bbox_to_anchor=(1.05, 1))
plt.tight_layout()
plt.show()
```

# Context Retrieval Impact

```python
plt.figure(figsize=(12, 6))
context_comparison = df.groupby('context_retrieval')[metrics].mean()
context_comparison.plot(kind='bar')
plt.title('Average Metrics using Context Retrieval')
plt.xticks(rotation=0)
plt.legend(bbox_to_anchor=(1.05, 1))
plt.tight_layout()
plt.show()
```

# Analysis of top best Configurations

```python
best_configs = pd.DataFrame({
    'Metric': metrics,
    'Best Value': [df[metric].max() for metric in metrics],
    'Best Configuration': [
        df.loc[df[metric].idxmax(), ['chunk_size', 'strategy', 'context_retrieval']].to_dict()
        for metric in metrics
    ]
})

print("\nBest Configurations by Metric:")
print(best_configs)
```
